```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_kosis=pd.read_csv("./data/population_kosis_1997_2020.csv",encoding="cp949")
df_kosis.shape
```

```python
#컬럼 전체출력 옵션설정
pd.options.display.max_columns = 865
```

```python
df_kosis.head()
```

```python
df_kosis.head(1)
```

```python
df=df_kosis.melt(id_vars="시군구별")
df.head()
```

```python
df["시군구별"].unique()
```

```python
print(df.shape)
#시군구별 데이터 제외
df=df[df["시군구별"] != "시군구별"].copy()
df.shape
```

```python
df["시군구별"].unique()
```

```python
df.head()
```

```python
df["연도"]=df["variable"].str.split(".",expand=True)[0]
df["월"]=df["variable"].str.split(".",expand=True)[1]
df["성별"]=df["variable"].str.split(".",expand=True)[2]
df.head()
```

```python
df.tail()
```

```python
df["성별"].unique()
```

```python
df["성별"].nunique()
```

```python
df["성별"]=df["성별"].fillna("전체")
df["성별"].unique()
```

```python
df["성별"]=df["성별"].replace("1","남").replace("2","여")
df["성별"].unique()
```

```python
df["성별"].value_counts()
```

```python
df.head()
```

```python
df = df.rename(columns={"variable": "기간", "value": "출생아수"})
df.head()
```

```python
df["출생아수"] = df["출생아수"].replace("-", np.nan)
df["출생아수"] = df["출생아수"].astype(float)
df["출생아수"].describe()
```

```python
df_all = df[(df["시군구별"] == "전국") & (df["성별"] == "전체")]
df_all.head(2)
```

```python
df_all=df_all[["연도", "월", "출생아수"]].copy()
df_all.head()
```

```python
plt.rc("font",family="Malgun Gothic")
df_all.set_index(["연도", "월"]).plot(figsize=(15,4))
```

```python
#최근 48개월 출생률
df_all[-48:].set_index(["연도", "월"]).plot.bar(figsize=(15,4))
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=df_all,x="연도",y="출생아수",ci=None)
```

```python
#월별 출생아 그래프
plt.figure(figsize=(15,4))
sns.lineplot(data=df_all,x="연도",y="출생아수",ci=None,hue="월")
```

```python
plt.figure(figsize=(15,4))
sns.barplot(data=df_all,x="연도",y="출생아수",ci=None)
```

```python
#지역별 출생률
df_local = df[df["시군구별"] !="전국"].copy()
df_local.head()
```

```python
plt.figure(figsize=(15,4))
sns.pointplot(data=df_local, x="연도",y="출생아수",hue="성별")
```

```python
df_local_all = df_local[df_local["성별"]=="전체"]
df_local_all.head()
```

```python
plt.figure(figsize=(15,4))
sns.pointplot(data=df_local_all,x="연도",y="출생아수")
```

```python
plt.figure(figsize=(12,4))
sns.pointplot(data=df_local_all,x="연도",y="출생아수",hue="시군구별",ci=None)
plt.legend(loc='center right',bbox_to_anchor=(1.18,0.5),ncol=1)
```

```python
#서울,경기도,세종시만 출생률그래프 그리기
df_local_2=df_local_all[df_local_all["시군구별"].isin(["서울특별시","경기도","세종특별자치시"])]
df_local_2.head()
```

```python
plt.figure(figsize=(15,4))
sns.pointplot(data=df_local_2,x="연도",y="출생아수",ci=None, hue="시군구별")
```
